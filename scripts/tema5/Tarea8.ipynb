### Calcula el porcentaje medio de variación del pulso por minuto entre antes y después de hacer ejercicio y compara el valor de los que hacen ejercicio habitualmente y los que no. ¿Observas mucha diferencia?

```python
def diff_pulso(x):
    x['diff.pulso'] = x['pulso.despues'] - x['pulso.antes']
    return x
df_diff_pulso = df.groupby('hace.deporte').apply(diff_pulso)
df_diff_pulso
df_diff_pulso.groupby('hace.deporte')['diff.pulso'].aggregate(['min', 'mean', 'max'])
df_diff_pulso.groupby('hace.deporte').aggregate({
    'diff.pulso': 'mean'
})
```

### otra manera de añadir columna

```python
df_variacion = df
df_variacion['variacion.pulso'] = abs(((df_variacion['pulso.despues'] - df_variacion['pulso.antes']) / df_variacion['pulso.antes']) * 100)
df_variacion.groupby('genero').mean().round(2)
    
```

### Calcula el porcentaje medio de variación del pulso por minuto entre antes y después de hacer ejercicio  para los estudiantes que hacen ejercicio habitualmente y compara el valor de los hombres con el de las mujeres. ¿Observas mucha diferencia?

```python
df_diff_pulso[(df_diff_pulso['hace.deporte'] == 'si')].groupby('genero')['diff.pulso'].aggregate(['min', 'mean', 'max'])
```

### Calcula el porcentaje medio de variación del pulso por minuto entre antes y después de hacer ejercicio para los estudiantes que no hacen ejercicio habitualmente y compara el valor de los fumadores con los no fumadores. ¿Observas mucha diferencia?

```python
df_diff_pulso[(df_diff_pulso['hace.deporte'] == 'no')].groupby('fuma')['diff.pulso'].aggregate(['min', 'mean', 'max'])
df_variacion[(df_variacion['hace.deporte'] == 'no')].groupby(['hace.deporte', 'fuma']).mean().round(2)
```
